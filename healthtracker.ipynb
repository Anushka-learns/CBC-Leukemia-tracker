# ðŸ©¸ CBC Blood Report Tracker for Leukemia Monitoring

This notebook analyzes blood test trends (Hemoglobin, Platelets, WBCs, Blasts) to monitor leukemia progression over time.

Based on sample CBC data from Febâ€“July 2025, it shows how key health indicators changed during diagnosis and treatment stages.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#load the Data set
df= pd.read_csv("CBC_Leukemia_Tracker.csv")
df['Date']= pd.to_datetime(df['Date'])
#Show first few rows
df.head()
#plot hemoglobin overtime
plt.figure(figsize=(8,4))

plt.plot(df['Date'], df['Hemoglobin (g/dL)'], marker='o', color='crimson')
plt.title("Hemoglobin levels overtime")
plt.xlabel("Date")
plt.ylabel("g/dL")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# WBC Count
plt.figure(figsize=(8,4))
plt.plot(df['Date'], df['WBC (x10^3/uL)'], marker='o', color='green')
plt.axhline(10, color='gray', linestyle='--', label='Normal Upper Limit')
plt.title("WBC Count Over Time")
plt.xlabel("Date")
plt.ylabel("x10Â³/uL")
plt.legend()
plt.grid(True)
plt.show()

# Blasts
plt.figure(figsize=(8,4))
plt.plot(df['Date'], df['Blasts (%)'], marker='o', color='black')
plt.title("Blast Percentage Over Time")
plt.xlabel("Date")
plt.ylabel("% of abnormal cells")
plt.grid(True)
plt.show()
```

## âœ… Key Insights:
- Hemoglobin dropped below normal after March, indicating anemia
- Platelet count sharply declined post-April, requiring transfusions
- WBCs and blast % rose dramatically in Juneâ€“July â†’ matching leukemia diagnosis
